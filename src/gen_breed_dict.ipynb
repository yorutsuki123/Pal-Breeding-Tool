```python
import pandas as pd
import json

df = pd.read_csv('breeding.csv')

breed_dict = [[[p, q], r] for p, q, r in df.values]
breed_dict += [[[q, p], r] for p, q, r in df.values]

with open('breed_dict_tmp.json', 'w') as f:
    json.dump(breed_dict, f)
```

```python
with open('../breed_dict.json') as f:
    data = json.load(f)
breed_dict = {(p, q): r for [p, q], r in data}

with open('breed_dict_tmp.json') as f:
    data = json.load(f)
breed_dict_tmp = {(p, q): r for [p, q], r in data}
```

```python
breed_dict_k = breed_dict.keys()
breed_dict_tmp_k = breed_dict_tmp.keys()

for i in breed_dict_tmp_k:
    if breed_dict[i] != breed_dict_tmp[i]:
        print(i, breed_dict[i], breed_dict_tmp[i])
```

```python
breed_dict_rev_tmp = {}
for p, q in breed_dict_tmp.items():
    if q not in breed_dict_rev_tmp:
        breed_dict_rev_tmp[q] = []
    breed_dict_rev_tmp[q].append(list(p))
with open('breed_dict_rev_tmp.json', 'w') as f:
    json.dump(breed_dict_rev_tmp, f)
```
